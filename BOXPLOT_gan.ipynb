```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the Excel file
file_path = '/content/synthetic_data (1).xlsx'
data = pd.read_excel(file_path)

# Combine input and output variables
columns = ['C (%)', 'H (%)', 'N (%)', 'O (%)', 'S (%)', 'FC (%)', 'Ash (%)', 'FC (%)',
           'Cel (%)', 'Hem (%)', 'Lig (%)']
data = data[columns]

# Melt the data for seaborn box plot
melted_data = data.melt(var_name='Variable', value_name='Value')

# Set custom color palette
palette = sns.color_palette("coolwarm", n_colors=len(columns))

# Create the fancy box plot
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid", font_scale=1.2)
ax = sns.boxplot(x='Variable', y='Value', data=melted_data, palette=palette)
ax.set_title('Fancy Box Plot of Input and Output Variables')

# Customize the legend
legend_patches = [plt.Line2D([], [], color=color, label=label, linewidth=0, marker='o', markersize=8)
                  for label, color in zip(columns, palette)]
ax.legend(handles=legend_patches, loc='upper right', title='Variables')

# Save the plot as an image file
plt.savefig('fancy_box_plot.png', dpi=300)

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the first Excel file
file_path1 = '/content/New_data_2 - Copy.xlsx'
data1 = pd.read_excel(file_path1)
data1['Dataset'] = 'Dataset 1'

# Read the second Excel file
file_path2 = '/content/synthetic_data (1).xlsx'
data2 = pd.read_excel(file_path2)
data2['Dataset'] = 'Dataset 2'

# Combine input and output variables
columns = ['C (%)', 'H (%)', 'N (%)', 'O (%)', 'S (%)', 'FC (%)', 'Ash (%)', 'FC (%)',
           'Cel (%)', 'Hem (%)', 'Lig (%)']

# Select the relevant columns and concatenate both datasets
data1 = data1[columns + ['Dataset']]
data2 = data2[columns + ['Dataset']]
combined_data = pd.concat([data1, data2], ignore_index=True)

# Melt the data for seaborn box plot
melted_data = combined_data.melt(id_vars=['Dataset'], var_name='Variable', value_name='Value')

# Set custom color palette
palette = sns.color_palette("coolwarm", n_colors=len(columns))

# Create the fancy box plot
plt.figure(figsize=(16, 6))
sns.set(style="whitegrid", font_scale=1.2)
ax = sns.boxplot(x='Variable', y='Value', hue='Dataset', data=melted_data, palette=palette)
ax.set_title('Fancy Box Plot Comparison of Input and Output Variables')

# Customize the legend and place it outside the main graph
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Datasets')

# Set axis labels
ax.set_xlabel('Variables')
ax.set_ylabel('Values')

# Save the plot as an image file
plt.savefig('fancy_box_plot_comparison.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the first Excel file (Original data)
file_path1 = '/content/New_data_2 - Copy.xlsx'
data1 = pd.read_excel(file_path1)
data1['Dataset'] = 'Original Data'

# Read the second Excel file (Synthetic data)
file_path2 = '/content/synthetic_data (1).xlsx'
data2 = pd.read_excel(file_path2)
data2['Dataset'] = 'Synthetic Data'

# Combine input and output variables
columns = ['C (%)', 'H (%)', 'N (%)', 'O (%)', 'S (%)', 'FC (%)', 'Ash (%)', 'FC (%)',
           'Cel (%)', 'Hem (%)', 'Lig (%)']


# Select the relevant columns and concatenate both datasets
data1 = data1[columns + ['Dataset']]
data2 = data2[columns + ['Dataset']]
combined_data = pd.concat([data1, data2], ignore_index=True)

# Melt the data for seaborn box plot
melted_data = combined_data.melt(id_vars=['Dataset'], var_name='Variable', value_name='Value')

# Set custom color palette
dataset_colors = sns.color_palette("Set2", n_colors=2)
palette = {dataset: color for dataset, color in zip(['Original Data', 'Synthetic Data'], dataset_colors)}

# Create the fancy box plot
plt.figure(figsize=(16, 6))
sns.set(style="whitegrid", font_scale=1.2)
ax = sns.boxplot(x='Variable', y='Value', hue='Dataset', data=melted_data, palette=palette)
ax.set_title('Fancy Box Plot Comparison of Input and Output Variables')

# Customize the legend and place it outside the main graph
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Datasets')

# Set axis labels
ax.set_xlabel('Variables')
ax.set_ylabel('Values')

# Save the plot as an image file
plt.savefig('fancy_box_plot_comparison.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the first Excel file (Original data)
file_path1 = '/content/New_data_2 - Copy.xlsx'
data1 = pd.read_excel(file_path1)

# Read the second Excel file (Synthetic data)
file_path2 = '/content/synthetic_data (1).xlsx'
data2 = pd.read_excel(file_path2)

# Calculate correlation matrices for both datasets
corr_data1 = data1.corr()
corr_data2 = data2.corr()

# Set up the figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8), sharey=True)

# Create the heatmap for the original data
sns.heatmap(corr_data1, cmap='coolwarm', annot=True, fmt='.2f', linewidths=.5, ax=ax1, cbar=False)
ax1.set_title('Heatmap of Original Data')

# Create the heatmap for the synthetic data
sns.heatmap(corr_data2, cmap='coolwarm', annot=True, fmt='.2f', linewidths=.5, ax=ax2, cbar_kws={'label': 'Correlation Coefficient'})
ax2.set_title('Heatmap of Synthetic Data')

# Save the plot as an image file
plt.savefig('heatmap_comparison.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
